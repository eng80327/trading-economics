```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Sample data
dates = ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01', '2023-05-01']
values = [10, 20, 15, 25, 30]

# Convert string dates to datetime objects
date_objects = [datetime.strptime(date, '%Y-%m-%d') for date in dates]

# Create a line plot
plt.plot(date_objects, values)

# Format x-axis date labels
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d, %Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Rotate x-axis date labels for better visibility
plt.gcf().autofmt_xdate()

# Set plot title and labels
plt.title('X-Axis Date Labels Formatting')
plt.xlabel('Date')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Index': [1, 2, 3, 4, 5],
    'Legend': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 15, 25, 30]
}

# Create a DataFrame
df = pd.DataFrame(data)
df.set_index('Index', inplace=True)

# Group the data by 'Legend'
grouped = df.groupby('Legend')

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot data for each group in separate subplots
for name, group in grouped:
    ax = axes[0] if name == 'A' else axes[1]
    ax.plot(group.index, group['Value'], label=name)
    ax.set_xlabel('Index')
    ax.set_ylabel('Value')
    ax.set_title(f'Group {name}')
    ax.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Index': [1, 2, 3, 4, 5],
    'Legend': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 15, 25, 30],
    'Value2': [5, 10, 7, 15, 18]
}

# Create a DataFrame
df = pd.DataFrame(data)
df.set_index('Index', inplace=True)

# Group the data by 'Legend'
grouped = df.groupby('Legend')

# Create a figure with a single plot and two y-axes
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot data for each group on the first y-axis
for name, group in grouped:
    ax1.plot(group.index, group['Value'], label=f'{name} - Value', linestyle='-', marker='o')
    ax1.set_xlabel('Index')
    ax1.set_ylabel('Value')
    ax1.set_title('Two Y-Axes with Vertically Stacked Subplots')
    ax1.legend()

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()

# Plot data for each group on the second y-axis
for name, group in grouped:
    ax2.plot(group.index, group['Value2'], label=f'{name} - Value2', linestyle='--', marker='x')
    ax2.set_ylabel('Value2')
    ax2.legend()

# Adjust layout and display the plot
fig.tight_layout()
plt.show()
```
