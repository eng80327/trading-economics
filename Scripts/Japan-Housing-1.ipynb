# Japan Housing

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_housing = pd.read_csv('../Data/Japan/3_Housing/historical_country_Japan_indicator_Residential_Property_Prices.csv')
df_housing['Date'] = pd.to_datetime(df_housing['DateTime'], format="%Y-%m-%dT%H:%M:%S")

df_housing.set_index(['Date'], inplace=True)
df_housing.sort_index(ascending=True)

df_housing['Value2'] = ((1 + df_housing['Value']/100)**(1/4) - 1)*100
df_housing['Value'] = df_housing['Value']/4

df_housing['CumValue'] = (1+df_housing['Value']/100).cumprod() - 1
df_housing['CumValue'] = df_housing['CumValue'] - df_housing.loc['1991-03-31', 'CumValue']

df_housing[(df_housing.index.year>=1991) & (df_housing.index.year < 2015)]
```

```python
fdf = df_housing[(df_housing.index.year>=1989) & (df_housing.index.year < 2012)]

fdf.Value.plot(figsize = (12,8))
plt.title("Japan Housing Index", fontsize = 15)
plt.legend(loc = 1, fontsize = 12)
plt.xlabel("Year", fontsize = 13)
plt.ylabel("Housing Index", fontsize = 13)

```

```python
fdf.CumValue.plot(figsize = (12,8))
plt.title("Japan Cumulative Housing Index", fontsize = 15)
plt.legend(loc = 1, fontsize = 12)
plt.xlabel("Year", fontsize = 13)
plt.ylabel("Housing Index", fontsize = 13)
```

## Insight about the Japanes Crash

- Takes about 20 years from 1991 to 2010 to bottom out.  It's like a slow painful correction
- Peak to Through for Japanese Property can be up to -40% to -50%
